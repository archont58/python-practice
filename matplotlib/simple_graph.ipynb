```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
x1 = range(0, 100)
y1 = [i+8 for i in x1]

x2 = range(0, 100, 5)
y2 = []
for i in range(20):
    y2.append(random.randint(0, 80))
    
x3 = [0, 20, 40, 60, 80, 100]
y3 = [41, 63, 25, 63, 20, 33]


    
plt.plot(x1, y1, color = 'red', linewidth = 3, linestyle = '--', label = 'linear')
plt.plot(x2, y2, color = 'green', linewidth = 2, label = 'random')
plt.plot(x3, y3, color = 'blue', linewidth = 3 , linestyle = ':', label = 'triangle')
plt.xlabel('x - axis')
plt.ylabel('y - label')
plt.title('Graph')
plt.legend()
plt.grid(True)
```
